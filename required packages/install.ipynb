```python
import subprocess as sp
import pandas as pd
```

```python
with open("pip_package.txt") as f:
    file = f.read().split("\n")
```

```python
pack = [line.split()[0] for line in file if line]
```

```python
template = "yes | pip install {}"
for p in pack:
    out = sp.run(template.format(p), shell=True)
    print(out)
```
